# Gradient Descent with Exact Line Search

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

$f: (x_1,x_2) \mapsto (x_1-2)^2 + 2(x_2+1)^2+1$

$\nabla f(x_1, x_2) = \begin{pmatrix} 2(x_1-2) \\ 4(x_2+1) \end{pmatrix} = \begin{pmatrix} 2x_1-4 \\ 4x_2+4 \end{pmatrix}$

$H_f(x_1, x_2) = \begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix}$

```python
x_1 = np.arange(-4,4,0.1)
x_2 = np.arange(-4,4,0.1)
X_1,X_2 = np.meshgrid(x_1,x_2)
```

```python
def f(x_1,x_2):
    return (x_1-2)**2 + 2*(x_2+1)**2 +1
def grad_f(x_1,x_2):
    return np.array([(2*x_1-4), (4*x_2+4)])
```

```python
X_3 = f(X_1, X_2)
fig=plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X_1,X_2,X_3,cmap=cm.viridis)
ax.set_xlabel("x1")
ax.set_ylabel("x2");
```

```python
plt.close()
fig,ax = plt.subplots()
ax.set_aspect(aspect=1)
CS = ax.contour(X_1,X_2,X_3, levels=[2,4.28,6,8,10,15,20,25,30,35,40])
ax.clabel(CS, inline=1, fontsize=10)
ax.plot(2,-1,"o")
ax.set_xlabel("x1")
ax.set_ylabel("x2");
```

```python
ax.plot([-4,2,2], [-1,-1,-4], "--");
```

$x^{(1)} = \begin{pmatrix} -2\\ 2 \end{pmatrix} + \lambda \cdot \begin{pmatrix} 8 \\-12 \end{pmatrix}$

$g (\lambda) = f(x^{(0)} + \lambda \cdot v) = f \left(\begin{pmatrix} -2\\ 2 \end{pmatrix} + \lambda \cdot \begin{pmatrix} 8 \\-12 \end{pmatrix}\right) = f(-2+8 \lambda, 2 -12 \lambda)$.

$x^{(1)} = \begin{pmatrix} -2\\ 2 \end{pmatrix} + 0.3 \cdot \begin{pmatrix} 8 \\ -12 \end{pmatrix} \approx \begin{pmatrix} 0.4 \\ -1.6 \end{pmatrix}$

```python
plt.close()
fig,ax = plt.subplots()
ax.set_aspect(aspect=1)
CS = ax.contour(X_1,X_2,X_3, levels=[2,4.28,6,8,10,15,20,26,30,35,40])
ax.plot(2,-1,"o")
ax.plot([-2,0.4,2],[2,-1.6,-4],'o--',markevery=[0,1])
ax.set_xlabel("x1")
ax.set_ylabel("x2");
```
